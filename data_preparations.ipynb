```python
df.isnull().sum()  # untuk cek missing values
```

## Menangani data yang kosong

```python
missing =  df.isnull().sum() / len(df) * 100 #menghitung persentase missing values
print(missing.sort_values(ascending=False) )
```

## Mengahpus kolom yang kosong dengan persentase missing value > 40%

```python
drop_cols = missing[missing > 40].index
df = df.drop(columns=drop_cols)
print(f"kolom yang dihapus: {drop_cols.tolist()}")
```

# Isi nilai kosong untuk kolom numerik dan kategorikal

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

# Numeric -> median
for col in num_cols:
    if df[col].isna().any():
        df[col] = df[col].fillna(df[col].median())

# Categorical -> mode
for col in cat_cols:
    if df[col].isna().any():
        mode_vals = df[col].mode(dropna=True)
        fill_val = mode_vals.iloc[0] if not mode_vals.empty else "Unknown"
        df[col] = df[col].fillna(fill_val)

# verifikasi
print((df.isna().sum() / len(df) * 100).sort_values(ascending=False).head(10))
```

# VISUALISASI PERBANDINGAN


```python
# Survived per Sex (bar)
ct = df.groupby('sex')['survived'].mean()  # rate selamat
plt.figure()
ct.plot(kind='bar')
plt.title('Rata-rata Survived per Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.tight_layout()
plt.show()
```

```python
# Survived per Pclass (bar)
ct2 = df.groupby('pclass')['survived'].mean()
plt.figure()
ct2.plot(kind='bar')
plt.title('Rata-rata Survived per Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.tight_layout()
plt.show()
```
